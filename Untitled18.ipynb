```python
SELECT
    t.num_automate,

    -- Infos référentiel
    r.code_banque,
    r.dab_hos_site,
    r.code_postal,
    r.longitude,
    r.latitude,

    -- Activité retraits
    SUM(CASE WHEN t.lib = 'RETRAIT' THEN 1 ELSE 0 END) AS nb_retraits,

    SUM(CASE
            WHEN t.lib = 'RETRAIT'
            THEN t.montant
            ELSE 0
        END) AS montant_total,

    AVG(CASE
            WHEN t.lib = 'RETRAIT'
            THEN t.montant
        END) AS montant_moyen_retrait,

    -- Captures carte
    SUM(CASE WHEN t.lib = 'CAPTURE_CARTE' THEN 1 ELSE 0 END) AS nb_captures_total,

    SUM(CASE
            WHEN t.lib = 'CAPTURE_CARTE'
             AND t.reseau = 'COS'
            THEN 1
            ELSE 0
        END) AS nb_captures_cos,

    -- Taux de capture
    CASE
        WHEN SUM(CASE WHEN t.lib = 'RETRAIT' THEN 1 ELSE 0 END) > 0
        THEN
            SUM(CASE WHEN t.lib = 'CAPTURE_CARTE' THEN 1 ELSE 0 END)
            /
            SUM(CASE WHEN t.lib = 'RETRAIT' THEN 1 ELSE 0 END)
        ELSE 0
    END AS taux_capture

FROM table_retrait t
LEFT JOIN table_ref_gab r
    ON t.num_automate = r.num_automate

GROUP BY
    t.num_automate,
    r.code_banque,
    r.dab_hos_site,
    r.code_postal,
    r.longitude,
    r.latitude
;
```
