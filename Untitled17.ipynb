```python
SELECT
    r.num_automate,

    -- total retraits 2025 (déjà calculé)
    SUM(r.nb_retraits) AS nb_retraits_2025,

    -- captures 2025
    COALESCE(c.nb_captures_2025, 0) AS nb_captures_2025,

    -- infos référentiel
    g.code_banque,
    g.dab_hos_site,
    g.code_postal,
    g.longitude,
    g.latitude,
    g.lib_site_implementation,
    g.type_gab_e_i

FROM TABLE_RETRAIT_2025 r

-- on calcule les captures à part
LEFT JOIN (
    SELECT
        num_automate,
        COUNT(1) AS nb_captures_2025
    FROM TABLE_BRUTE
    WHERE lib = 'CAPTURE_CARTE'
      AND annee = 2025     -- adapte si besoin
    GROUP BY num_automate
) c
ON r.num_automate = c.num_automate

-- jointure référentiel
LEFT JOIN TABLE_REF_GAB g
ON r.num_automate = g.num_automate

GROUP BY
    r.num_automate,
    c.nb_captures_2025,
    g.code_banque,
    g.dab_hos_site,
    g.code_postal,
    g.longitude,
    g.latitude,
    g.lib_site_implementation,
    g.type_gab_e_i
;
```
