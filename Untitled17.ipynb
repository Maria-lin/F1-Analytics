```python
import dataiku
import pandas as pd

# ==============================
# LOAD DATA
# ==============================

vop = dataiku.Dataset("TABLE_VOP").get_dataframe()
roues = dataiku.Dataset("LISTE_ROUES").get_dataframe()

# ==============================
# CLEAN KEY
# ==============================

vop["identifiant_interne"] = vop["identifiant_interne"].astype(str).str.strip()
roues["identifiant_interne"] = roues["identifiant_interne"].astype(str).str.strip()

# ==============================
# JOIN (filtrer les bonnes roues)
# ==============================

df = vop.merge(
    roues[["identifiant_interne"]],
    on="identifiant_interne",
    how="inner"
)

print("nb lignes filtr√©es:", len(df))

# ==============================
# DETAIL DATASET
# ==============================

detail = df.copy()

# ==============================
# ANALYSES
# ==============================

# par statut
statut = (
    df.groupby("statut")
    .size()
    .reset_index(name="nb_requetes")
    .sort_values("nb_requetes", ascending=False)
)

statut["type"] = "statut"
statut.rename(columns={"statut":"valeur"}, inplace=True)

# par motif
motif = (
    df.groupby("motif")
    .size()
    .reset_index(name="nb_requetes")
    .sort_values("nb_requetes", ascending=False)
)

motif["type"] = "motif"
motif.rename(columns={"motif":"valeur"}, inplace=True)

# par circuit
circuit = (
    df.groupby("circuit")
    .size()
    .reset_index(name="nb_requetes")
    .sort_values("nb_requetes", ascending=False)
)

circuit["type"] = "circuit"
circuit.rename(columns={"circuit":"valeur"}, inplace=True)

# concat analyse
analyse = pd.concat([statut, motif, circuit])

# ==============================
# CROISE (optionnel mais top)
# ==============================

croise = (
    df.groupby(["statut","motif","circuit"])
    .size()
    .reset_index(name="nb_requetes")
    .sort_values("nb_requetes", ascending=False)
)

# ==============================
# WRITE OUTPUT
# ==============================

dataiku.Dataset("VOP_ROUES_DETAIL").write_with_schema(detail)
dataiku.Dataset("VOP_ROUES_ANALYSE").write_with_schema(analyse)
dataiku.Dataset("VOP_ROUES_CROISE").write_with_schema(croise)

```
